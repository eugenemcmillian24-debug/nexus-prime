```
import time
import random

def simulate_deployment(app_name, files):
    print(f"🚀 NEXUS PRIME // DEPLOYMENT ENGINE // INITIATING: {app_name.upper()}")
    print("-" * 60)
    
    steps = [
        ("INITIALIZING", "Fetching build environment..."),
        ("PROVISIONING", "Allocating Edge Runtime resources..."),
        ("CLONING", f"Syncing {len(files)} files to deployment node..."),
        ("BUILDING", "Next.js 14 Build optimization (Static Generation)..."),
        ("LINTING", "Type-safety validation (TSC)..."),
        ("OPTIMIZING", "Purging CSS and minifying JS bundles..."),
        ("UPLOADING", "Streaming assets to global CDN..."),
        ("VERIFYING", "Health check: nexus-sky-prod-v1.vercel.app [200 OK]")
    ]
    
    for step, detail in steps:
        time.sleep(0.5)
        print(f"[{step:12}] {detail}")
        if step == "BUILDING":
            print("  > ✓ Compiled 42 modules")
            print("  > ✓ Created 12 static routes")
        if step == "OPTIMIZING":
            print(f"  > ✓ Bundle size: {random.randint(120, 180)}KB (Optimal)")

    print("-" * 60)
    print(f"✅ DEPLOYMENT SUCCESSFUL: {app_name}")
    print(f"🔗 LIVE URL: https://{app_name.lower().replace(' ', '-')}.vercel.app")
    print("-" * 60)

simulate_deployment("Nexus Sky", ["app/page.tsx", "tailwind.config.js", "package.json"])
```
